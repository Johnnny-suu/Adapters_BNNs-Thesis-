```python
BNN_results = pd.read_csv('Binary_Resnet.csv',names = ['epoch','loss','accuracy'],header = None)
Baseline = pd.read_csv('Non_binary_model.csv',names = ['epoch','loss','accuracy'],header = None)
```

```python
BNN_results.head(), Baseline.head()
```

```python
#Best Accuracy

for df,name in zip([BNN_results.accuracy,Baseline.accuracy],['BNN','Baseline']):
    print(f'Best accuracy for {name}: {df.max()*100:.1f}%')
    
print(f'Difference in accuracy = {abs(BNN_results.accuracy.max() - Baseline.accuracy.max())*100:.1f}%')
```

```python


bnn = BNN_results
base = Baseline

fig,ax1 = plt.subplots(figsize = (10,6))
ax2 = ax1.twinx()

#losses
l1= ax1.plot('epoch','loss','m--',data = bnn  , label = 'BNN loss')
l2=ax1.plot('epoch','loss','b--',data = base, label = 'baseline loss')
ax1.set_ylabel('Loss')
ax1.set_xlabel('epoch')

#Accuracy
a1= ax2.plot('epoch','accuracy','m-',data = bnn, label = 'BNN acuraccy')
a2 =ax2.plot('epoch','accuracy','b-',data = base, label = 'baseline acuraccy')
ax2.set_ylim(0.3,1)
ax2.set_ylabel('Accuraccy')

lns = l1+l2+a1+a2
labs = [l.get_label() for l in lns]
ax1.legend(lns, labs,bbox_to_anchor=(0., -0.2, 1., .102), loc='upper left',mode = 'expand',ncol=4)
plt.title('Training of BNN vs Baseline Resnet18 Network on Cifar 10')



plt.show()
```
